```python
#importing SBopen class from mplsoccer to open the data
from mplsoccer import Sbopen, Pitch, VerticalPitch

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# The first thing we have to do is open the data. We use a parser SBopen available in mplsoccer.
parser = Sbopen()
#opening data using competition method
df_competition = parser.competition()
#opening data using match method
df_euros = parser.match(competition_id=55, season_id=282)
```

```python
df_netherlands = df_euros[(df_euros['home_team_country_name'] == 'Netherlands') | (df_euros['away_team_country_name'] == 'Netherlands')]
```

```python
netherlands_match_ids = df_netherlands['match_id'].unique()
```

```python
# Create a dictionary with keys as 'match_1', 'match_2', etc., and values as the event DataFrames
dfs_netherland = []

for match_id in netherlands_match_ids:
    df_event, df_related, df_freeze, df_tactics = parser.event(match_id)
    dfs_netherland.append(df_event)
df_netherlands = pd.concat(dfs_netherland)

    
```

```python
from matplotlib.cm import get_cmap

# Create the pitch
pitch = Pitch(line_color='black')
fig, ax = pitch.grid(grid_height=0.9, title_height=0.06, axis=False,
                     endnote_height=0.04, title_space=0, endnote_space=0)

# Filter for 'Ball Receipt' events
mask_gakpo_receipts = (gakpo_events.type_name == 'Ball Receipt')
df_gakpo_receipts = gakpo_events.loc[mask_gakpo_receipts, ['x', 'y', 'match_id', 'team_name']]

# Get unique match_ids and their corresponding team names
unique_matches = df_gakpo_receipts[['match_id', 'team_name']].drop_duplicates()

# Create a colormap (e.g., 'tab20' for up to 20 unique colors)
cmap = get_cmap('tab20')

# Create a list for legend handles
legend_handles = []

# Loop through the unique match IDs and team names
for i, (match_id, team_name) in enumerate(unique_matches.itertuples(index=False)):
    # Filter the DataFrame for the current match_id
    df_match = df_gakpo_receipts[df_gakpo_receipts['match_id'] == match_id]
    
    # Assign a color from the colormap based on the match_id index
    color = cmap(i / len(unique_matches))  # Normalize the color index to the number of unique matches
    
    # Plot the ball receipt locations for this match_id
    for _, row in df_match.iterrows():
        pitch.scatter(120 - row.x, 80 - row.y, alpha=0.6, s=100, color=color, ax=ax['pitch'])
    
    # Add a legend handle for this match
    legend_handles.append(plt.Line2D([0], [0], marker='o', color=color, label=f'{team_name}', 
                                     markersize=10, linestyle='None'))

# Add a legend for the matches
ax['pitch'].legend(handles=legend_handles, title="Matches", loc='upper right', fontsize=12)

# Set the title and show the plot
fig.suptitle("Gakpo Ball Receipts by Match", fontsize=30)
plt.show()
```
